```python
# Only use the following columns from the dataframe
clean_reduced_df = clean_df[["Series", "Player_1", "Player_2", "Rank_1", "Rank_2", "Winner"]]
clean_reduced_df
```

```python
rf_cr_df = clean_reduced_df[clean_reduced_df.Winner == 'Federer R.']
rf_cr_df
```

```python
# Shows just the times Federer won where he was listed as Player 1
rf_cr_p1_df = rf_cr_df[rf_cr_df.Player_1 == 'Federer R.']
rf_cr_p1_df

play1_wins_df = len(rf_cr_p1_df)
type(play1_wins_df)
```

```python
# Shows just the times Federer won where he was listed as Player 2
rf_cr_p2_df = rf_cr_df[rf_cr_df.Player_2 == 'Federer R.']
rf_cr_p2_df

play2_wins_df = len(rf_cr_p2_df)
type(play2_wins_df)
```

```python
# simple pie chart of wins as player 1 vs wins as player 2
# tell matplotlib to create a pie chart based upon the 
my_data = [play1_wins_df, play2_wins_df]
my_labels = ['Player 1', 'Player 2']
plt.pie(my_data, labels=my_labels,autopct='%1.1f%%', startangle=140)
plt.title('Roger Federer\nWins as Player 1 VS \nWins as Player 2')
plt.axis('equal')
plt.show()
```

```python
# Plot a double line graph tracking the player 1 and player 2 rankings for Roger Federer's matches since 2000
plt.plot(rf_cr_df['Rank_1'], label='Rank of Player 1', color='green',linewidth=0.3)
plt.plot(rf_cr_df['Rank_2'], label='Rank of Player 2', color='steelblue', linewidth=0.3)

#add legend
plt.legend()

#display plot
plt.show()
```

```python
# Find the outliers when Rank_1 is higher than 300
rf1_outliers = rf_cr_df[rf_cr_df.Rank_1 >= 300]
rf1_outliers
# clean_reduced_df[clean_reduced_df.Winner == 'Federer R.']
```

```python
# Find the outliers when Rank_2 is higher than 300
rf2_outliers = rf_cr_df[rf_cr_df.Rank_2 >= 300]
rf2_outliers
```

```python
rf_over_300_df = pd.merge(rf1_outliers,rf2_outliers,how='outer')
rf_over_300_df
```

```python
# Focus on John Isner, who is from Greensboro, NC and whose professional tennis career started in 2007
isner1_cr_df = clean_reduced_df[clean_reduced_df.Player_1 == 'Isner J.']
isner2_cr_df = clean_reduced_df[clean_reduced_df.Player_2 == 'Isner J.']
isner_cr_df = pd.merge(isner1_cr_df, isner2_cr_df, how='outer')
isner_cr_df
```

```python
# HEatmap
# Drop unneeded columns (Player_1, Player_2, Winner)
isner_df = isner_cr_df.drop(columns=['Player_1', 'Player_2'])
isner_series_df = isner_cr_df.groupby(["Series"]).count()
print(isner_series_df)
sns.heatmap(isner_series_df, linewidths = 0.30, annot = True)
```
